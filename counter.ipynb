# NOTE

To run this notebook, you will need to run the following command to extract the
data from the ReCom-A BEN file:

```bash
ben -m decode ./square_A_1M.jsonl.ben -v
```

This will create a file called `square_A_1M.jsonl` which contains the decoded
data and will take up ~680MB of disk space.

````python
import jsonlines as jl
import numpy as np
from tqdm.notebook import tqdm
````

````python
import ast
````

````python
with open("square_A_1M_changed_assignments.txt") as f:
    datars = f.read()
    datars = ast.literal_eval(datars)

datars = np.array(datars)
datars
````

````python
with jl.open("square_A_1M.jsonl") as f:
    line = f.read()
    curr_assignment = np.array(line["assignment"])
    count_flips = np.zeros(len(curr_assignment))
    for line in tqdm(f, total = 1_000_000):
        new_assignment = np.array(line["assignment"])
        flips_array = (curr_assignment != new_assignment)
        if np.any(flips_array):
            count_flips += flips_array.astype(int)
            curr_assignment = new_assignment
````
